## 6) Implement the Checker (Binary Pass/Fail)

You now need to implement the **checker function**, which will **automatically evaluate** whether the model’s answer is correct.

The function should be called:

```python
check_prediction(pred: dict, expected: dict) -> bool
```

and it must return:

* `True` → if the model’s answer is **exactly correct** or satisfies all defined rules.
* `False` → if the model’s answer is incorrect, incomplete, or violates any rule.

Your checker can handle one of two verification styles:

1. **Exact-match checking** — compares the model’s output directly to the expected answer.
2. **Rule-based checking** — validates logical or numerical properties (e.g., numeric tolerance, constraint satisfaction, symbolic equivalence).

All checkers must be:

* **Deterministic** — same input → same result every time.
* **Binary** — must output only pass/fail (True/False).
* **Context-aware** — the correct output depends on the Gold Context.

````python
def check_prediction(pred: dict, expected: dict) -> bool:
    try:
        return (
            pred.get("species") ==
            expected["answer"]["species"]
            and
            pred.get("region") ==
            expected["answer"]["region"]
            and
            pred.get("approved_quota") ==
            expected["answer"]["approved_quota"]
        )
    except:
        return False

# Check if working
# sample_pred = {
#     "species": "SilverHeron",
#     "region": "Coastal Zone A",
#     "approved_quota": 86
# }

# print(check_prediction(sample_pred, expected))
````
